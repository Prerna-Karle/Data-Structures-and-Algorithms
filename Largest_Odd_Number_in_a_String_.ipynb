```python
class Solution:
  def largeOddNum(self, s:str) -> str:
    ind = -1

    i = 0
    for i in range(len(s) -1, -1, -1):
      if(int(s[i])%2) == 1:
        ind = i
        break

    i = 0
    while i<= ind and s[i] == '0':
      i += 1

    return s[i:ind+1]

solution = Solution()
num = "504"
result = solution.largeOddNum(num)
print("Largest odd number:", result)
```

```python
# Leetcode

class Solution:
    def largestOddNumber(self, num: str) -> str:
        ind = -1

        # Find the rightmost odd digit
        for i in range(len(num) - 1, -1, -1):
            if int(num[i]) % 2 == 1:
                ind = i
                break

        # Remove leading zeros
        i = 0
        while i <= ind and num[i] == '0':
            i += 1

        return num[i:ind + 1]
```
