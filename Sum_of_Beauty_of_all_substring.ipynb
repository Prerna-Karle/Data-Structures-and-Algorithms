```python
# Approach

def beauty_sum(s):
  n = len(s)
  total = 0

  for i in range(n):
    freq = {}

    for j in range(i,n):
      freq[s[j]] = freq.get(s[j],0) + 1

      values = freq.values()
      maxi = max(values)
      mini = min(values)

      total += (maxi - mini)

    return total

def main():
  s = "xyx"
  print("Beauty Sum:", beauty_sum(s))

if __name__ == "__main__":
  main()
```

```python
# Leetcode

class Solution:
    def beautySum(self, s: str) -> int:
        n = len(s)
        total = 0

        # Consider every substring
        for i in range(n):
            freq = {}

            for j in range(i, n):
                # Update frequency of current character
                freq[s[j]] = freq.get(s[j], 0) + 1

                # Find maximum and minimum frequencies
                maxi = max(freq.values())
                mini = min(freq.values())

                # Add beauty of current substring
                total += (maxi - mini)

        return total
```
