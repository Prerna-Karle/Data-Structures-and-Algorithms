#Brute Force Approach

```python
def isSorted(arr,n):
  for i in range(n):
    for j in range(i + 1, n):
      if arr[i] > arr[j]:
         return False
  return True

arr = [1,2,3,4,5]
n = len(arr)
ans = isSorted(arr,n)
print(ans)
```

#Optimal approach

```python
def isSorted(arr,n):
  for i in range(1,n):
    if arr[i] < arr[i - 1]:
      return False
    return True

arr = [1,2,3,4,5]
n = len(arr)

print("True" if isSorted(arr,n) else "False")
```

#Another one

```python
def isSorted(arr,n):
  for i in range(0, n-1):
    if arr[i] > arr[i+1]:
      return False
  return True

arr = [1,2,3,4,5]
n = len(arr)

ans = isSorted(arr,n)
print(ans)
```
