```python
# Brute-Force Approach

class Solution:
  def floorSqrt(self, n: int) -> int:
    ans = 0

    for i in range(1, n+1):
      if i * i <= n:
        ans = i
      else:
        break

    return ans

n = 27
sol = Solution()
print(sol.floorSqrt(n))
```

```python
# Optimal Approach

class Solution:
  def mySqrt(self, x: int) -> int:
    if x < 2:
      return x

    left, right, ans = 1, x // 2, 0

    while left <= right:
      mid = (left + right) // 2

      if mid * mid <= x:
        ans = mid
        left = mid + 1
      else:
        right = mid - 1

    return ans

if __name__ == "__main__":
  s = Solution()
  print(s.mySqrt(8))
```
