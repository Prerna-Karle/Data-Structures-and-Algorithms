```python
nums = [55, 32, -97, 99, 3, 67]

largest = nums[0]
n = len(nums)

for i in range(n):
    largest = max(largest, nums[i])

print(largest)
```

```python
# Alternative

nums = [55, 32, -97, 99, 3, 67]

largest = float('-inf')
n = len(nums)

for i in range(n):
  largest = max(largest, nums[i])

print(largest)
```
