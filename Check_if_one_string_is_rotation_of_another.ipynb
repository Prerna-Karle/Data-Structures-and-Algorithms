```python
# Brute force approach

class Solution:
  def rotateString(self, s: str, goal: str) -> bool:
    if len(s) != len(goal):
      return False

    for i in range(len(s)):
      rotated = s[i:] + s[:i]
      if rotated == goal:
        return True

    return False

sol = Solution()
print(sol.rotateString("rotation", "tionrota"))
```

```python
# Optimal approach

class Solution:
  def rotateString(self, s: str, goal: str) -> bool:
    if len(s) != len(goal):
      return False
    doubled_s = s + s
    return goal in doubled_s

sol = Solution()
print(sol.rotateString("rotation", "tionrota"))
```

```python
# Leetcode

class Solution:
    def rotateString(self, s: str, goal: str) -> bool:

        # Strings must have the same length
        if len(s) != len(goal):
            return False

        # Concatenate s with itself
        doubled_s = s + s

        # Check if goal is present in doubled_s
        return goal in doubled_s
```
