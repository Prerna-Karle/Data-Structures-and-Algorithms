```python
# Approach

class Solution:
  def frequencySort(self, s):
    freq = [(0, chr(i + ord('a'))) for i in range(26)]

    for ch in s:
      index = ord(ch) - ord('a')
      freq[index] = (freq[index][0] + 1, ch)

    freq.sort(key=lambda x: (-x[0], x[1]))

    result = [ch for f, ch in freq if f > 0]
    return result

if __name__ == "__main__":
  sol = Solution()
  s = "tree"
  result = sol.frequencySort(s)
  print(result)
```

```python
# Leetcode

class Solution:
    def frequencySort(self, s: str) -> str:

        # Dictionary to store the frequency of each character
        freq = {}

        # Count the frequency of each character
        for ch in s:
            freq[ch] = freq.get(ch, 0) + 1

        # Sort characters by frequency in descending order
        sorted_characters = sorted(
            freq.keys(),
            key=lambda ch: -freq[ch]
        )

        # Create the final string
        result = ""

        # Add every character according to its frequency
        for ch in sorted_characters:
            result += ch * freq[ch]

        # Return the final string
        return result
```
