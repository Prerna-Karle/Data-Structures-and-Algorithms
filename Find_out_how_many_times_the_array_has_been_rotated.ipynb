```python
# Brute Force Approach

class Solution:
    def findRotations(self, arr):
       n = len(arr)
       minVal = arr[0]
       minIndex = 0

       for i in range(1, n):
        if arr[i] < minVal:
          minVal = arr[i]
          minIndex = i

       return minIndex

if __name__ == "__main__":
  obj = Solution()
  arr = [4, 5, 6, 7, 0, 1, 2, 3]
  rotations = obj.findRotations(arr)
  print(rotations)
```

```python
# Better Approach

class Solution:
  def findRotationCount(self, arr):
    n = len(arr)
    for i in range(n - 1):
      if arr[i] > arr[i + 1]:
        return i + 1
    return 0

if __name__ == "__main__":
  arr = [3,4,5,1,2]
  sol = Solution()
  rotations = sol.findRotationCount(arr)
  print(rotations)
```

```python
# Optimal Approach

class Solution:
  def findRotationCount(self, arr):
    low = 0
    high = len(arr) - 1

    while low < high:
      mid = low + (high - low) // 2
      if arr[mid] > arr[high]:
        low = mid + 1
      else:
        high = mid
    return low

if __name__ == "__main__":
  arr = [4,5,6,7,0,1,2,3]
  sol = Solution()
  rotations = sol.findRotationCount(arr)
  print(rotations)
```
