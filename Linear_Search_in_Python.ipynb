```python
nums = [5,3,9,8,1,6,4,-10,-100]
target = 4

def linearSearch(nums, target):
 for i in range(len(nums)):
    if nums[i] == target:
       return i
 return -1

result = linearSearch(nums, target)
print(result)
```
