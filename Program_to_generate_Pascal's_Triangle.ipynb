```python
# Approach-1

class Solution:
  def generate(self, numRows):
    triangle = []

    for i in range(numRows):
      row = [1] * (i + 1)

      for j in range(1, i):
        row[j] = triangle[i - 1][j - 1] + triangle[i - 1][j]

      triangle.append(row)

    return triangle

if __name__ == "__main__":
  obj = Solution()
  n = 5

  result = obj.generate(n)
  for row in result:
    print(" ".join(map(str, row)))
```

```python
# Approach 2

class Solution:
  def getNthRow(self, N):
    row = []

    val = 1
    row.append(val)

    for k in range(1, N):
      val = val * (N - k) // k
      row.append(val)

    return row

N = 5
sol = Solution()
result = sol.getNthRow(N)

print(*result)
```

```python
# Approach - 3

class Solution:
  def findPascalElement(self, r, c):
    n = r - 1
    k = c - 1

    result = 1

    for i in range(k):
      result *= (n - i)
      result //= (i + 1)

    return result

if __name__ == "__main__":
  sol = Solution()
  r, c = 5, 3
  print(sol.findPascalElement(r,c))
```
