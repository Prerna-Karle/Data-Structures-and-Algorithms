```python
# Brute Force Approach

class Solution:
    def findMin(self, nums):
      min_val = float('inf')
      for i in range(len(nums)):
        min_val = min(min_val, nums[i])
      return min_val

nums = [4, 5, 6, 7, 0, 1, 2]
sol = Solution()
result = sol.findMin(nums)
print("Minimum element is", result)
```

```python
# Optimal Approach

class Solution:
  def findMin(self, nums):
    low, high = 0, len(nums) - 1
    while low < high:
      mid = low + (high - low) // 2
      if nums[mid] > nums[high]:
        low = mid + 1
      else:
        high = mid
    return nums[low]

nums = [4, 5, 6, 7, 0, 1, 2]
sol = Solution()
result = sol.findMin(nums)
print("Minimum element is", result)
```

```python
# Leetcode

from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        low, high = 0, len(nums) - 1

        while low < high:
            mid = low + (high - low) // 2

            if nums[mid] > nums[high]:
                low = mid + 1
            else:
                high = mid

        return nums[low]
```
