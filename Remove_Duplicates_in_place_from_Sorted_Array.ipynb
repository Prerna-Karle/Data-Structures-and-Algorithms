#Brute force approach

```python
# Using Dictionary

nums = [1,1,1,2,3,4,4,7,9,9,9,10]

n = len(nums)
freq_map = {} #Empty dictionary

for i in range(0, n):
  freq_map[nums[i]] = 0

j = 0

for k in freq_map:
  nums[j] = k
  j += 1

print("Number of unique elements:", j)
print("Modified array:", nums[:j])
```

```python
# Using sets

class Solution:
    def removeDuplicates(self, nums):
        seen = set()
        index = 0

        for num in nums:
            if num not in seen:
                seen.add(num)
                nums[index] = num
                index += 1

        return index

nums = [1, 1, 1, 2, 3, 4, 4, 7, 9, 9, 9, 10]

sol = Solution()
k = sol.removeDuplicates(nums)

print("k =", k)
print("Array after removing duplicates:", nums[:k])
```

#Optimal Solution

```python
class Solution:
    def removeDuplicates(self, nums):
        n = len(nums)
        if n == 1:
            return 1

        i = 0
        j = i + 1

        while j < n:
            if nums[j] != nums[i]:
                i += 1
                nums[i], nums[j] = nums[j], nums[i]

            j += 1

        return i + 1

nums = [1, 1, 2, 2, 3, 4, 4, 7, 9, 9]

sol = Solution()

k = sol.removeDuplicates(nums)

print("k =", k)
print("Array after removing duplicates:", nums[:k])
```
