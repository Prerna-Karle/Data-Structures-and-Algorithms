```python
# Brute Force Approach

class Solution:
  def rotateClockwise(self, matrix):
    n = len(matrix)

    rotated = [[0] * n for _ in range(n)]

    for i in range(n):
      for j in range(n):
        rotated[j][n - i - 1] = matrix[i][j]

    return rotated

matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

obj = Solution()
rotated = obj.rotateClockwise(matrix)

for row in rotated:
  print(row)
```

```python
# Optimal Solution

class Solution:
  def rotateClockwise(self, matrix):
    n = len(matrix)

    # Step 1 : Transpose the matrix
    for i in range(n):
      for j in range(i + 1, n):
        matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

    # Step 2 : Reverse each row
    for i in range(n):
      matrix[i].reverse()

matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

obj = Solution()
obj.rotateClockwise(matrix)

for row in matrix:
  print(*row)
```

```python
# Leetcode

from typing import List

class Solution:
  def rotate(self, matrix: List[List[int]]) -> None:
    n = len(matrix)

    # Step 1 : Transpose the matrix
    for i in range(n):
      for j in range(i, n):
        matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

    # Step 2 : Reverse each row
    for row in matrix:
      row.reverse()
```
