#Brute Force Approach

```python
class Solution:
    # Function to find the element that appears only once
    def getSingleElement(self, arr):
        n = len(arr)

        # Check each element's count
        for i in range(n):
            num = arr[i]
            count = 0

            # Count how many times num appears
            for j in range(n):
                if arr[j] == num:
                    count += 1

            # Return if it appears exactly once
            if count == 1:
                return num

        return -1

arr = [4, 1, 2, 1, 2]
obj = Solution()
ans = obj.getSingleElement(arr)
print("The single element is:", ans)
```

#Better Approach

```python
class Solution:
  def getSingleElement(self, arr):
    n = len(arr)

    max_val = max(arr)

    hash_arr = [0] * (max_val + 1)

    for num in arr:
      hash_arr[num] += 1

    for num in arr:
      if hash_arr[num] == 1:
        return num

    return -1

arr = [4, 1, 2, 1, 2]
obj = Solution()
ans = obj.getSingleElement(arr)
print("The single element is:", ans)
```

#Optimised Approach

```python
class Solution:
  def getSingleElement(self, arr):
    xorr = 0

    for num in arr:
      xorr ^= num

    return xorr

arr = [4, 1, 2, 1, 2]
obj = Solution()
ans = obj.getSingleElement(arr)
print("The single element is:", ans)
```
