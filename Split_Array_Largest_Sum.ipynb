```python
# Brute Force Approach

class SubarrayPartitioner:
    def count_partitions(self, a, max_sum):
      partitions = 1
      subarray_sum = 0

      for num in a:
        if subarray_sum + num <= max_sum:
          subarray_sum += num
        else:
          partitions += 1
          subarray_sum = num

      return partitions

    def largest_subarray_sum_minimized(self, a, k):
      low = max(a)
      high = sum(a)

      for max_sum in range(low, high + 1):
        if self.count_partitions(a, max_sum) == k:
          return max_sum

      return low

if __name__ == "__main__":
  a = [10, 20, 30, 40]
  k = 2
  sp = SubarrayPartitioner()
  ans = sp.largest_subarray_sum_minimized(a, k)
  print("The answer is:", ans)
```

```python
# Optimal Approach

class SubarrayPartitioner:
  def count_partitions(self, a, max_sum):
    partitions = 1
    subarray_sum = 0

    for num in a:
      if subarray_sum + num <= max_sum:
        subarray_sum += num
      else:
        partitions += 1
        subarray_sum = num
    return partitions

  def largest_subarray_sum_minimized(self, a, k):
    low = max(a)
    high = sum(a)

    while low <= high:
      mid = (low + high) // 2
      partitions = self.count_partitions(a, mid)

      if partitions > k:
        low = mid + 1
      else:
        high = mid - 1
    return low

if __name__ == "__main__":
  a = [10, 20, 30, 40]
  k = 2
  sp = SubarrayPartitioner()
  ans = sp.largest_subarray_sum_minimized(a, k)
  print("The answer is:", ans)
```

```python
# Leetcode

class Solution:
    def countPartitions(self, nums, maxSum):
        partitions = 1
        subarraySum = 0

        for num in nums:
            if subarraySum + num <= maxSum:
                subarraySum += num
            else:
                partitions += 1
                subarraySum = num

        return partitions

    def splitArray(self, nums, k):
        low = max(nums)
        high = sum(nums)

        while low <= high:
            mid = (low + high) // 2

            partitions = self.countPartitions(nums, mid)

            if partitions > k:
                low = mid + 1
            else:
                high = mid - 1

        return low
```
