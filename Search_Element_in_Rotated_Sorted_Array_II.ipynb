```python
# Brute Force Approach

class Solution:
    def search_in_rotated_sorted_array_II(self, arr, k):
      for num in arr:
        if num == k:
          return True
      return False

if __name__ == "__main__":
  arr = [7, 8, 1, 2, 3, 3, 3, 4, 5, 6]
  k = 3
  sol = Solution()
  result = sol.search_in_rotated_sorted_array_II(arr, k)

  if result:
    print("Target is present in the array.")
  else:
    print("Target is not present.")
```

```python
# Optimised Approach

class Solution:
  def search_in_rotated_sorted_array_II(self, arr, k):
    low, high = 0, len(arr) - 1

    while low <= high:
      mid = (low + high) // 2

      if arr[mid] == k:
        return True

      if arr[low] == arr[mid] == arr[high]:
        low += 1
        high -= 1
        continue

      if arr[low] <= arr[mid]:
        if arr[low] <= k <= arr[mid]:
          high = mid - 1
        else:
          low = mid + 1
      else:
        if arr[mid] <= k <= arr[high]:
          low = mid + 1
        else:
          high = mid - 1

    return False

if __name__ == "__main__":
  arr = [7, 8, 1, 2, 3, 3, 3, 4, 5, 6]
  k = 3
  sol = Solution()
  result = sol.search_in_rotated_sorted_array_II(arr, k)

  if result:
    print("Target is present in the array.")
  else:
    print("Target is not present.")
```

```python
# Leetcode

class Solution:
    def search(self, nums: list[int], target: int) -> bool:

        low, high = 0, len(nums) - 1

        while low <= high:

            mid = (low + high) // 2

            # Target found
            if nums[mid] == target:
                return True

            # Duplicates: cannot determine sorted half
            if nums[low] == nums[mid] == nums[high]:
                low += 1
                high -= 1
                continue

            # Left half is sorted
            if nums[low] <= nums[mid]:

                if nums[low] <= target < nums[mid]:
                    high = mid - 1
                else:
                    low = mid + 1

            # Right half is sorted
            else:

                if nums[mid] < target <= nums[high]:
                    low = mid + 1
                else:
                    high = mid - 1

        return False
```
