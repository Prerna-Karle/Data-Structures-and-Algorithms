```python
# Brute Force Approach

class MissingKFinder:
  def missing_k(self, vec, k):
    for num in vec:
      if num <= k:
        k += 1
      else:
        break

    return k

vec = [4,7,9,10]
k = 4

finder = MissingKFinder()
ans = finder.missing_k(vec, k)

print('The missing number is:', ans)
```

```python
# Optimal approach

class MissingKFinder:
  def missing_k(self, vec, k):
    low, high = 0, len(vec) - 1

    while low <= high:
      mid = (low + high) // 2

      missing = vec[mid] - (mid+1)

      if missing < k:
        low = mid + 1
      else:
        high = mid - 1

    return k + high + 1

vec = [4, 7, 9, 10]
k = 4

finder = MissingKFinder()
ans = finder.missing_k(vec, k)

print("The missing number is:", ans)
```

```python
# Leetcode

from typing import List

class Solution:
    def findKthPositive(self, arr: List[int], k: int) -> int:
        low, high = 0, len(arr) - 1

        while low <= high:
            mid = (low + high) // 2

            # Missing numbers before arr[mid]
            missing = arr[mid] - (mid + 1)

            if missing < k:
                low = mid + 1
            else:
                high = mid - 1

        return k + high + 1
```
