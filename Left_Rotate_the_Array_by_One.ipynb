```python
def leftRotateByOne(arr,n):
  temp=arr[0]
  for i in range(n-1):
    arr[i]=arr[i+1]
  arr[n-1]=temp
  return arr

arr=[1,2,3,4,5]
print(leftRotateByOne(arr,len(arr)))
```

#Brute Force

```python
def solve(arr,n):
  temp = [0] * n

  # shift elements to the left
  for i in range(1,n):
    temp[i-1] = arr[i]

  # put the first element at the end
  temp[n-1] = arr[0]

  # print the rotated array
  for num in temp:
    print(num, end=" ")
  print()

if __name__=="__main__":
  n = 5
  arr = [1,2,3,4,5]
  solve(arr,n)
```

#Optimal approach

```python
class Solution:
    def rotateArrayByOne(self, nums):
        temp = nums[0]

        # Shift elements to the left
        for i in range(1, len(nums)):
            nums[i-1] = nums[i]

        # Put first element at the end
        nums[-1] = temp


if __name__ == "__main__":
    solution = Solution()
    nums = [1, 2, 3, 4, 5]

    solution.rotateArrayByOne(nums)
    print(nums)
```
