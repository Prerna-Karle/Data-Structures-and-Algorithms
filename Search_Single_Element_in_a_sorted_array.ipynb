```python
# Brute force Approach 1

class Solution:
  def singleNonDuplicate(self,arr):
    n = len(arr)
    if n == 1:
      return arr[0]
    for i in range(n):
      if i == 0:
        if arr[i] != arr[i+1]:
          return arr[i]
      elif i==n-1:
        if arr[i] != arr[i-1]:
          return arr[i]
      else:
        if arr[i] != arr[i-1] and arr[i] != arr[i+1]:
          return arr[i]
    return -1

if __name__ == "__main__":
  arr = [1,1,2,2,3,3,4,5,5,6,6]
  obj = Solution()
  ans = obj.singleNonDuplicate(arr)
  print("The single element is:", ans)
```

```python
# Brute Approach 2

class Solution:
  def singleNonDuplicate(self,arr):
    n = len(arr)
    ans = 0

    for i in range(n):
      ans ^= arr[i]
    return ans

if __name__ == "__main__":
  arr = [1,1,2,2,3,3,4,5,5,6,6]
  obj = Solution()
  ans = obj.singleNonDuplicate(arr)
  print("The single element is:", ans)
```

```python
# Optimal Approach

class Solution:
  def singleNonDuplicate(self,arr):
    n = len(arr)

    if n == 1:
      return arr[0]

    if arr[0] != arr[1]:
      return arr[0]

    if arr[n-1] != arr[n-2]:
      return arr[n-1]

    low, high = 1, n-2

    while low <= high:
      mid = (low+high)//2

      if arr[mid] != arr[mid+1] and arr[mid] != arr[mid-1]:
        return arr[mid]

      if (mid % 2 == 1 and arr[mid] == arr[mid - 1]) or \
          (mid % 2 == 0 and arr[mid] == arr[mid + 1]):
          low = mid + 1
      else:
        high = mid - 1

    return -1

if __name__ == "__main__":
  arr = [1,1,2,2,3,3,4,5,5,6,6]
  obj = Solution()
  ans = obj.singleNonDuplicate(arr)
  print("The single element is:", ans)
```

```python
# Leetcode

from typing import List

class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        low, high = 0, len(nums) - 1

        while low < high:
            mid = (low + high) // 2

            if nums[mid] == nums[mid ^ 1]:
                low = mid + 1
            else:
                high = mid

        return nums[low]
```

```python
from typing import List

class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        n = len(nums)

        # Only one element
        if n == 1:
            return nums[0]

        # First element is unique
        if nums[0] != nums[1]:
            return nums[0]

        # Last element is unique
        if nums[n - 1] != nums[n - 2]:
            return nums[n - 1]

        low, high = 1, n - 2

        while low <= high:
            mid = (low + high) // 2

            # Found the unique element
            if nums[mid] != nums[mid - 1] and nums[mid] != nums[mid + 1]:
                return nums[mid]

            # We are on the correct side of the pairing pattern
            if (mid % 2 == 1 and nums[mid] == nums[mid - 1]) or \
               (mid % 2 == 0 and nums[mid] == nums[mid + 1]):
                low = mid + 1
            else:
                high = mid - 1

        return -1
```
