```python
# Approach

class BinarySearchInsert:
  def search_insert(self, arr, x):
    n = len(arr)
    low, high = 0, n - 1
    ans = n

    while low <= high:
      mid = (low + high) // 2

      if arr[mid] >= x:
        ans = mid
        high = mid - 1
      else:
        low = mid + 1
    return ans

if __name__ == "__main__":
  arr = [1, 2, 4, 7]
  x = 6
  obj = BinarySearchInsert()
  index = obj.search_insert(arr, x)
  print(f"The index is: {index}")
```

```python
# Leetcode

class Solution:
  def searchInsert(self, nums: list[int], target: int) -> int:

    low = 0
    high = len(nums) - 1

    ans = len(nums)

    while low <= high:
      mid = (low + high)//2
      if nums[mid] >= target:
        ans = mid
        high = mid - 1
      else:
        low = mid + 1
    return ans
```
