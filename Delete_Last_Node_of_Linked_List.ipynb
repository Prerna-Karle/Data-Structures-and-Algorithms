```python
# Approach

class Node:
  def __init__(self, val):
    self.data = val
    self.next = None

class Solution:
  def deleteTail(self, head):
    if head is None or head.next is None:
      return None

    curr = head
    while curr.next.next is not None:
      curr = curr.next

    curr.next = None

    return head

head = Node(1)
head.next = Node(2)
head.next.next = Node(3)

obj = Solution()
head = obj.deleteTail(head)

temp = head
while temp:
  print(temp.data, end=" ")
  temp = temp.next
```
